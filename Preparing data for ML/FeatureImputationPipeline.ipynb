```python
import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
```

```python
diabetes=pd.read_csv('C:/Users/shubham.chhetri/Desktop/MYPROJECT/Machine learning/Preparing data for ML/diabetes_process_complete.csv')
```

```python
diabetes.head()
```

```python
diabetes_features=diabetes.drop(['Unnamed: 0','Outcome'],axis=1)
diabetes_label=diabetes[['Outcome']]
```

```python
mask=np.random.randint(0,100,size=diabetes_features.shape).astype(np.bool)#random make it 0
mask=np.logical_not(mask)
```

```python
diabetes_features[mask]=np.nan
diabetes_features.sample(15)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train,x_test,y_train,y_test=train_test_split(diabetes_features,diabetes_label,test_size=0.2)
```

```python
transformer=ColumnTransformer(transformers=[('features',SimpleImputer(strategy='mean'),[0,1,2,3,4,5,6,7])])
```

```python
clf=make_pipeline(transformer,DecisionTreeClassifier(max_depth=4))
clf=clf.fit(x_train,y_train)
clf.score(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```
