```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans
```

```python
data=pd.read_csv('C:/Users/shubham.chhetri/Desktop/MYPROJECT/Machine learning/Creating ML models/Mall_Customers.csv',index_col=0)
```

```python
data.head()
```

```python
data.drop(['Genre','Age'],axis=1,inplace=True)
data.head()
```

```python
data=data.sample(frac=1)#shuffle
```

```python
k_means=KMeans(n_clusters=2)
k_means.fit(data)
```

```python
k_means.labels_
```

```python
np.unique(k_means.labels_)
```

```python
centers=k_means.cluster_centers_
centers
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'],c=k_means.labels_,s=100)
plt.scatter(centers[:,0],centers[:,1],color='blue',marker='s',s=200)
plt.xlabel('Annual Income')
plt.ylabel('Spending Income')
plt.title('K-Means with 2 clusters ')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
score=silhouette_score(data,k_means.labels_)
print("score",score)
```
